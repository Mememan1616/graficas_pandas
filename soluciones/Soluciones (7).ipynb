1.	Lee en dos dataframes `gdp` y `unemployment`, el contenido de los archivos GDP.csv y Unemployment.csv respectivamente, que almacenan por país, el PIB por persona empleada y la tasa de desempleo desde 1991.  

```python
gdp = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TLG/DS_C3_SC2_T4_Pruebate/GDP.csv")
gdp
```

```python
unemployment = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TLG/DS_C3_SC2_T4_Pruebate/Unemployment.csv")
unemployment
```

2.	A partir de los dataframes anteriores obtén uno (`info2019`) donde sólo almacenes la información de ambos indicadores en el año 2019. Haz los cambios necesarios para que quede con la siguiente estructura.

```python
info2019 = gdp[['Country Code','2019']].merge(unemployment[['Country Code','2019']],on='Country Code')
info2019.columns = ['Country Code','GDP','Unemployment']
info2019
```

```python
sns.set_palette('Spectral')
sns.set_style('ticks')
```

3.	Grafica el dataframe anterior, utilizando la plataforma Matplotlib, para visualizar la distribución de estos índices a través de histogramas en ejes duales. Personaliza las propiedades del gráfico para que se vea como sigue: 

```python
fig,ax1 = plt.subplots() 
ax1.hist(x='GDP',data=info2019,alpha=0.5,label='GDP') 
ax1.set_xlabel('GDP')
ax1.set_ylabel('Frecuency')

ax2 = ax1.twiny() 
ax2.hist(x='Unemployment',data=info2019,color='darkseagreen',alpha=0.5,label='Unemployment')
ax2.set_xlabel('Unemployment')

plt.title('World 2019')
fig.legend(bbox_to_anchor=(0.86,0.87))
plt.show()
```

4.	Almacena en un dataframe (`continents`), el contenido del archivo Continents.csv, que indica a qué contienente y región pertenece cada país y combínalo con el dataframe `info2019`, usando la columna Country Code, para obtener el dataframe `continents_2019`

```python
continents = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TLG/DS_C3_SC2_T4_Pruebate/Continents.csv")
continents
```

```python
continents_2019 = info2019.merge(continents,on='Country Code')
continents_2019
```

5.	Del dataframe anterior, obtén los promedios por continente de ambos índices. Almacena el resultado en `continents_group`

```python
continents_group = continents_2019.groupby(['Continent']).mean()
continents_group
```

6.	Crea una matriz en la que vayas añadiendo las siguientes subgráficas:

*   Subgráfica 1
  *   Tipo de gráfico: Scatter
  *   Objetivo: Relación entre los índices GDP y Unemployment
  *   dataframe: `info2019`
  *   Plataforma: Pandas

*   Subgráfica 2
  *   Tipo de gráfico: Boxplot
  *   Objetivo: Dispersión de la tasa de desempleo por continente
  *   dataframe: `continents2019`
  *   Plataforma: Seaborn

*   Subgráfica 3
  *   Tipo de gráfico: Circular
  *   Objetivo: Comparar el  GDP promedio por continente
  *   dataframe: `continents_group`
  *   Plataforma: Pandas

```python
fig = plt.figure(figsize=(17,5)) 
plt.subplots_adjust(wspace=0.4)

ax1 = fig.add_subplot(131)
info2019.plot(kind='scatter',x='GDP',y='Unemployment',color='indianred',logx=True,ax=ax1)
ax1.set_title('World 2019')

ax2 = fig.add_subplot(132)
sns.boxplot(x='Continent',y='Unemployment',data=continents_2019,ax=ax2)
ax2.set_title('Unemployment by continent 2019')

ax3 = fig.add_subplot(133)
continents_group.plot.pie(y='GDP',labels=continents_group.index,autopct='%1.1f%%',ax=ax3)
ax3.set_title('GDP average by continent 2019')
ax3.set_ylabel('')

plt.show()
```

7. Almacena la información de México en un dataframe con la estructura mostrada y grafica ambos índices para visualizar la evolución a través de los años utilizando ejes duales y Seaborn

```python
mexico = pd.concat([gdp[gdp['Country Name']=='Mexico'], unemployment[unemployment['Country Name']=='Mexico']], axis=0)
mexico.drop(['Country Name','Country Code'],axis=1,inplace=True)
mexico = mexico.T
mexico.index.name = 'Year'
mexico.columns.name = ''
mexico.columns = ['GDP','Unemployment']
mexico
```

```python
ax1 = sns.lineplot(x=mexico.index,y=mexico['GDP'],label='GDP')
ax1.set_title('Mexico')

ax2 = ax1.twinx()
sns.lineplot(x=mexico.index,y=mexico['Unemployment'],ax=ax2,color='darkseagreen',label='Unemployment')

plt.xticks(np.arange(0,len(mexico)+1,4))
plt.show()
```
